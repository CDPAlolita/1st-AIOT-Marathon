```python
import pymongo
```

```python
client = pymongo.MongoClient(host='127.0.0.1', port=27017)
```

```python
db = client['Member']
collection = db['Member']
```

資料庫情況:


```python
#插入一筆資料
mydate = {'name':'Andy','email':'andy@email.com.tw','phone':'0912345678'}
result = collection.insert_one(mydate)
print(result.inserted_id)
```

資料庫情況:


```python
#新增多筆
data_list = [
    {'name':'Bob','email':'bob@email.com.tw','phone':'0912345677'},
    {'name':'Cindy','email':'cindy@email.com.tw','phone':'0912345676'},          
    {'name':'Dan','email':'dan@email.com.tw','phone':'0912345675'},
    {'name':'Error','email':'error@email.com.tw','phone':'0912345678999'}
]
result = collection.insert_many(data_list)
print(result.inserted_ids)
```

資料庫情況:


```python
#刪除資料
collection.delete_one({'name':'Error'})
```

資料庫情況:
    

```python
#修改所有符合條件資料
filter_ = {'name':'Andy'}
update_ = {"$set": {'name':'Amy'}}
collection.update_many(filter_,update_)
```

```python
#查詢更改後的結果
result = collection.find()
for i in result:
    print(i)
```
